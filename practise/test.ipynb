```python
import pandas as pd
import numpy as np
import datetime
```

# CPU_Util_.csv

```python
df = pd.read_csv('CPU_Util_.csv')

df.head()
```

```python
df['TimeStamp'] = pd.to_datetime(df['Time (GMT)'])
```

```python
df['Hr'] = [x. hour for x in df['TimeStamp']]
```

```python
df.head()
```

```python
df.groupby('Hr').mean()
```

```python
 df.groupby('Hr').min()
```

```python
df.groupby('Hr').max()
```
